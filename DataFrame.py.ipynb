2) Create DataFrame

```python
list1=["Ram","Pratik","Rushi"]
list2=[30,32,33]
names=pd.DataFrame(zip(list1,list2),
             columns=["Name","Age"])
names
```

```python
# What Name You Want To Provide : List.csv
# Step 1: Save it in Colab
names.to_csv("List.csv", index=False) # Index = False Means It Will Not Create Index Column

# Download From Colab To Local
from google.colab import files
files.download("List.csv")
```
